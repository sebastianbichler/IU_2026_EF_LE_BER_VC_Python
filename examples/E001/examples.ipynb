```python
my_string = "Hello World"
print(my_string[0::3])
```

```python
my_set1 = set()
my_set2 = {1, 2, 3, 4, 5, 5, 5, 5, 5}
my_set1.add(10)
my_set1.add(20)
for item in my_set2:
    print(item * 2)
```

```python
my_set3 = {1, 'a', 3.14, (1, 2)}
print(my_set3)

for element in my_set3:
    print(type(element), ": ", element)
```

```python
my_tuple = (10, 20, 30, 40, 50)
print("Länge des Tupels:", len(my_tuple))
for value in my_tuple:
    print(value + 5)
```

```python
my_list = [100, 200, 300, 400, 500]
print("Länge der Liste:", len(my_list))
for value in my_list:
    print(value - 10)
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
print("Länge des Dictionaries:", len(my_dict))
for key, value in my_dict.items():
    print(key, ":", value * 3)
```

## Matplotlib Beispiel

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, marker="o")
plt.title("Einfacher Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
