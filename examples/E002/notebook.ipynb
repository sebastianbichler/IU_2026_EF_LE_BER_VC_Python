```python
import pandas as pd
# Online-Quelle:
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
# Als lokale Sicherung speichern
df.to_csv("daten.csv", index=False)
```

```python
with open("daten.csv", "r") as f:
    header = f.readline() # Header Ã¼berspringen
    daten = []
    for line in f:
        spalten = line.strip().split(",")
        # Datentyp-Konvertierung (Cast)
        value = float(spalten[0])
        daten.append(value)

durchschnitt = sum(daten) / len(daten)
print(f"Datentyp: {type(daten)}, Durchschnitt: {durchschnitt}")
```

```python
table = []

with open("daten.csv", "r") as f:
    header = f.readline().strip().split(",") # Namen der Spalten merken
    number_cols = len(header)

    for row in f:
        values = row.strip().split(",")
        # Konvertierung: Versuche Zahlen in floats zu wandeln, sonst bleib beim String
        converted_row = []
        for value in values:
            try:
                converted_row.append(float(value))
            except ValueError:
                converted_row.append(value)

        table.append(converted_row)

# infos
number_rows = len(table)
print(f"Table has {number_rows} rows and {number_cols} columns.")
for colValue in table:
    print(f"Data type row: {type(colValue)} and its value is {colValue[0]}")
```

```python
import numpy as np
np_daten = np.array(daten)
# Operation auf alle Elemente gleichzeitig (Broadcasting)
normalisierte_daten = (np_daten - np_daten.mean()) / np_daten.std()
print(f"NumPy Durchschnitt: {np_daten.mean()}")
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("iris.sqlite")
query = "SELECT * FROM daten WHERE species = 'setosa'"
df_setosa = pd.read_sql(query, conn)
print(df_setosa.head())

query = "SELECT AVG(sepal_length) AS avg_sepal_length FROM daten WHERE species = 'setosa'"
avg_sepal_length = pd.read_sql(query, conn).iloc[0, 0] # Access the value
print(f"\nAverage sepal_length for setosa: {avg_sepal_length:.3f}") # if you do not use .iloc, you get a DataFrame and need to format it differently with [0]
conn.close()
```
