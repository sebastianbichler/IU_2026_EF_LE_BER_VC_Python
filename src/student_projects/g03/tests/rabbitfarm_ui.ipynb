Wir müssen Bezug auf unsere bestehenden Models.py nehmen und daher den Pfad ordentlich deklarieren
Außerdem macht es Sinn den DATA Pfad miteinzubeziehen wo unsere JSON Datei liegt.

Widgets Informationen hier : https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html

```python
import ipywidgets as widgets
from ipywidgets import Button, Layout
from IPython.display import display
```

```python
Headline = widgets.Label(value='Max kleine Rabbitfarm', 
                        disabled=False, 
                        style=dict(
                            font_style='italic',
                            font_weight='bold',
                            font_variant="small-caps",
                            text_color='green',
                            text_decoration='underline',
                            text_size='40px'
))
output = widgets.Output()
display(Headline, output)
GemueseButton = widgets.Button(description="Gemüseverwaltung", layout=Layout(width='16%', height='80px'))
GemueseButton.style.button_color = 'lightgreen'
display(GemueseButton)
def on_button_clicked(b):
    with output:
        print("Gemüseverwaltung gestartet!")

GemueseButton.on_click(on_button_clicked)
```

```python
from ipywidgets import Button, HBox, VBox
output = widgets.Output()

headline = widgets.Label(value='Max kleine Rabbitfarm', 
                        disabled=False,
                        grid_column='span 4',
                        justify_self='stretch', 
                        style=dict(
                            font_style='italic',
                            font_weight='bold',
                            font_variant="small-caps",
                            text_color='green',
                            text_decoration='underline',
                            text_size='40px'
))

words = ['Gemüseverwaltung', 'Beeteverwaltung', 'Lagerbestände', 'Kundenverwaltung', 'Einnahmen/Ausgaben', 'Sensordaten', 'Beenden']
items = [Button(description=w, style=dict(
    button_color='lightblue')) for w in words]
left_box = VBox([items[0], items[1]])
center_box = VBox([items[2], items[3]])
right_box = VBox([items[4], items[5]])
exit = VBox([items[6]])  
display(Headline, HBox([left_box, center_box,right_box, exit]), output)

def on_button_clicked(b):
    output.clear_output()
    with output:
        print(f"{b.description} gestartet!")

for item in items:
    item.on_click(on_button_clicked)
```
