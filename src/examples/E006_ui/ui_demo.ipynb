## IPyWidgets Example

```python
from ipywidgets import interact

def pengu_restaurant(fisch, menge):
    print(f"Pingu serviert {menge} Portion(en) {fisch}!")

# Erstellt automatisch ein Dropdown und einen Slider!
interact(pengu_restaurant, fisch=['Hering', 'Lachs', 'Sardine'], menge=(1, 10));
```

```python
import ipywidgets as widgets
from IPython.display import display

# 1. Controls erstellen
fisch_dropdown = widgets.Dropdown(
    options=['Hering', 'Lachs', 'Thunfisch'],
    description='Fischart:',
)

menge_slider = widgets.IntSlider(
    value=1, min=1, max=20,
    description='Menge:'
)

bestell_button = widgets.Button(
    description='Bestellung aufgeben',
    button_style='success' # grüner Button
)

ausgabe = widgets.Output() # Bereich für Rückmeldungen

# 2. Logik definieren
def verarbeite_bestellung(b):
    with ausgabe:
        ausgabe.clear_output()
        print(f"✅ Bestellung aufgenommen: {menge_slider.value}x {fisch_dropdown.value}")

bestell_button.on_click(verarbeite_bestellung)

# 3. UI anzeigen (Layout durch Boxen)
ui = widgets.VBox([fisch_dropdown, menge_slider, bestell_button, ausgabe])
display(ui)
```
