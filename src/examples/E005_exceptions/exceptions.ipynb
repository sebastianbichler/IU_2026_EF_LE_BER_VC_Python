## Exception handling try/except

```python
print("Program start")
a = 10
b = int(input("Enter a divisor: "))
try:
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("You tried to divide by zero. " + "Please use a non-zero divisor.")
print("Program end")
```

```python
try:
    user_input = input("Enter an integer divisor: ")
    divisor = int(user_input)
    result = 100 / divisor
    print("Result:", result)
except ValueError:
    print("That was not a valid integer.")
except ZeroDivisionError:
    print("Divisor must not be zero.")
```

```python
try:
    user_input = input("Enter an integer: ")
    value = int(user_input)
except ValueError:
    print("Not a valid integer.")
else:
    print("You entered:", value)
```

```python
try:
    user_input = input("Enter an integer: ")
    user_div_input = input("Enter an integer divisor: ")
    divisor = int(user_div_input)
    # my_variable = 10 / 10
    my_variable = int(user_input) / divisor
except ZeroDivisionError:
    print("Divided by zero!")
finally:
    print("Here\’s our finally block...")
print("End of program")
```

```python
try:
    my_variable = 10 / 0
except ZeroDivisionError:
    print("Divided by zero!")
finally:
    print("Here\’s our finally block...")
print("End of program")
```

## Reading a File Safely

```python
try:
    file_name = input("Enter file name: ")
    my_file = open(file_name, "r")
    print(my_file.read())
    my_file.close()
except FileNotFoundError:
    print("File not found")
print("End of program")
```

## Raising exceptions yourself

```python
def set_player_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    print("Age set to", age)


set_player_age(25)
set_player_age(-3)
```

```python
try:
    raise ZeroDivisionError
finally:
    print("Here\’s our finally block...")
print("End of program")
```

```python
try:
    raise ZeroDivisionError
except ZeroDivisionError:
    print("Divided by zero!")
finally:
    print("Here\’s our finally block...")
print("End of program")
```
