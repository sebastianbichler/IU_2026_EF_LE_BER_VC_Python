### Writing text to file

```python
my_file = open("../E007_soccer/players.txt", "w")
my_file.write("Alex,73\n")
my_file.write("Morgan,68\n")
my_file.close()
```

### Reading text from file

```python
my_file = open("../E007_soccer/players.txt", "r")
content = my_file.read()
print(content)
my_file.close()
```

```python
with open("../E007_soccer/players.txt", "r") as f:
    for line in f:
        name, weight = line.strip().split(",")
        print(name, "weighs", weight, "kg")
```
