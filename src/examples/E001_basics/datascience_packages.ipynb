## NumPy 

```python
import numpy as np

data = np.array([1, 2, 3, 4, 5])
mean_value = np.mean(data)
squared = data**2

print(data)
print("Mean: ", mean_value)
print("Squared values: ", squared)
```

## pandas

```python
import pandas as pd

df = pd.DataFrame({"player": ["Alice", "Bob", "Charlie"], "goals": [10, 7, 5]})

print(df["goals"].mean())
```

## Matplotlib

```python
import matplotlib.pyplot as plt

players = ["Alice", "Bob", "Charlie"]
goals = [10, 7, 5]

plt.bar(players, goals)
plt.title("Goals per Player")
plt.xlabel("Player")
plt.ylabel("Goals")
plt.show()

plt.scatter(players, goals)
plt.show()
```
