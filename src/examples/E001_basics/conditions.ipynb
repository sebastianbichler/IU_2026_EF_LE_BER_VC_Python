## Example conditions

```python
a = int(input("Gib eine Zahl ein a ="))
b = int(input("Gib eine Zahl ein b ="))

if a>b:
    print("a ist größer als b.")
else: 
    print("a ist nicht größer als b.")
```

## Multiple Conditions

```python
a = int(input("Gib eine Zahl ein a ="))
b = int(input("Gib eine Zahl ein b ="))

if a>b:
    print("a ist größer als b.")
elif a<b:
    print("a ist kleiner als b.")
else: 
    print("a ist gleich b.")
```

## loops

```python
for x in range(10):
    if x == 3:
       break
    print(x)
```

```python
for x in range(10):
    if x == 3:
       continue
    print(x)
```

```python
## !! infinite loop
x = 0
while(x<10):
    if x == 3:
       continue
    print(x)
    x+=1
```

```python
x = 0
while(x<10):
    x+=1
    if x == 3:
       continue
    print(x)
 
```
