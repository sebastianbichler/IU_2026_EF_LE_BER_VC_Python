**Basic Line Charts**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.01)
y = np.tan(x)
plt.plot(x, y)
plt.show()
```

**Simple Sine Example** 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 500)  # x values (0-2)
y = np.sin(x)

# plot configuration
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="sin(x)", color="blue")
plt.title("Simple Sine Curve")
plt.xlabel("x (Radiant)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```
