### Exkurs Laufzeitmodell (Variable, Instanz)

```python
my_name = "John"
your_name = my_name
print(my_name, your_name)
print(id(my_name), id(your_name))
my_name = "Peter"
print(my_name, your_name)
print(id(my_name), id(your_name))
```
