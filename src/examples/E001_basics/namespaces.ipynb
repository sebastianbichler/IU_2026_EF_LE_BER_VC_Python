### Namespaces: redefine floor

```python
from math import floor

print(floor(6.5))

def floor(a):
    if a == 1:
        return "First floor"
    elif a == 2:
        return " Second floor"
    elif a == 3:
        return "Third floor"

print(floor(2))
```

### avoid conflicts 

```python
import math

print(math.floor(6.5))

def floor(a):
    if a == 1:
        return "First floor"
    elif a == 2:
        return " Second floor"
    elif a == 3:
        return "Third floor"

print(floor(2))
```

### Global an local namespaces

```python
my_str = "Hello, World!"
print(my_str)

def print_it_x_times(text, x):
    a = 0
    while a < x:
        print(text)
        a += 1

print_times = 5
print_it_x_times("Hello again, World!", print_times)
```
