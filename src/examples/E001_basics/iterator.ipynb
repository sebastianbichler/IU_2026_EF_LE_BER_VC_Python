## Iterators

```python
my_list = [4, 8, 15, 16, 23, 42]
my_iterator = iter(my_list)

print(next(my_iterator)) # 4
print(next(my_iterator)) # 8
print(next(my_iterator)) # 15
print(next(my_iterator)) # 16
print(next(my_iterator)) # 23
print(next(my_iterator)) # 42
```

```python
my_list = [4, 8, 15, 16, 23, 42]
for value in my_list:
    print(value)
```

## List comprehensions

```python
# squares of numbers
numbers = [1, 2, 3, 4, 5]
squares = [n * n for n in numbers]
print(squares)
```

```python
my_list = list(range(11))
my_new_list = [x**2 for x in my_list if x%2 == 0]
print(my_list)
print(my_new_list)
```

```python
# Filter Even Numbers
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = [n for n in numbers if n % 2 == 0]
print(even_numbers) 
```
