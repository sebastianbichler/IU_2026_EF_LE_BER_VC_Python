## For-Schleife

```python
my_players = ["Franz Beckenbauer", "Gerd Muller", "Lothar Matthaus", "Manuel Neuer"]
for x in range(4):
    print("Player " + str(x + 1) + ": " + my_players[x])
```

## While-Schleife

```python
my_players = ["Franz Beckenbauer", "Gerd Muller", "Lothar Matthaus", "Manuel Neuer"]

x = 0
while x < 4:
    print("Player " + str(x + 1) + ": " + my_players[x])
    x += 1

print("End of roster")
```

## Skipping Player

```python
my_players = ["Franz Beckenbauer", "Gerd Muller", "Lothar Matthaus", "Manuel Neuer"]
for x in range(4):
    if x == 2:
        continue
    print("Player " + str(x + 1) + ": " + my_players[x])
```

```python
my_players = ["Franz Beckenbauer", "Gerd Muller", "Lothar Matthaus", "Manuel Neuer"]
for x in range(4):
    if x == 2:
        break
    print("Player " + str(x + 1) + ": " + my_players[x])
```

## Comprehension with String

```python
my_players = ["Franz Beckenbauer", "Gerd Muller", "Lothar Matthaus", "Manuel Neuer"]
last_names = [name.split()[1] for name in my_players]

print(last_names)
```

## Exporting Player Data

```python
players = {"Alex": 73, "Morgan": 68, "Taylor": 80}

with open("players.csv", "w") as f:
    for name, weight in players.items():
        f.write(f"{name},{weight}\n")

new_players = {}
with open("players.csv", "r") as f:
    for line in f:
        name, weight = line.strip().split(",")
        new_players[name] = weight
        print(new_players.keys())
        print(new_players.values())

for name, weight in new_players.items():
    print(name, "weighs", weight, "kg.")
```

## Collections - Lists

```python
players = ["Alex", "Morgan", "Taylor"]
print(players[0])  # "Alex"

players.append("Jordan")
players[1] = "Casey"

print(players)
```

```python
numbers = [5, 2, 9, 1]
print(len(numbers))  # 4
print(min(numbers))  # 1
print(max(numbers))  # 9
numbers.sort()
print(numbers)  # [1, 2, 5, 9]
# Slicing works on lists as on strings:
print(numbers[1:3])
```

## Collections - Dictionaries

```python
player_weights = {"Alex": 73, "Morgan": 68, "Taylor": 80}
print(player_weights["Alex"])  # 73

player_weights["Jordan"] = 75  # add new entry
player_weights["Alex"] = 74  # update existing
print(player_weights.keys())
print(player_weights.values())

# Looping:
for name, weight in player_weights.items():
    print(name, "->", weight)
```

## Collections - Sets

```python
a = {"Alex", "Morgan", "Taylor"}
b = {"Morgan", "Jordan"}
print(a & b)  # intersection
print(a | b)  # union
print(a - b)  # difference
print(b - a)  # difference
```
