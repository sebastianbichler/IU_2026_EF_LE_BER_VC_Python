## Tkinter Basics (the simplest GUI library for Python)

```python
import tkinter as tk
from tkinter import messagebox


def on_click():
    name = entry.get().strip() or "Welt"
    messagebox.showinfo("Hallo", f"Hallo, {name}!")


root = tk.Tk()
root.title("Simple GUI")

tk.Label(root, text="Name:").grid(row=0, column=0, padx=8, pady=8, sticky="w")
entry = tk.Entry(root, width=25)
entry.grid(row=0, column=1, padx=8, pady=8)

tk.Button(root, text="Begrüßen", command=on_click).grid(
    row=1, column=0, columnspan=2, padx=8, pady=8
)

root.mainloop()
```
